```python
# ***

# An alternativoe to using youtube-dl is to use the pytube library
# We can use the pytube library to download the video and then extract the audio from the video

!pip install pytube
```

```python
# We import the pytube library

from pytube import YouTube

# We use the url variable that we set earlier to create a YouTube object

yt = YouTube('https://www.youtube.com/watch?v=iYYRH4apXDo')

# We can use the get_audio_only() method to get the audio stream

stream = yt.streams.get_audio_only()

# Then we get the audio and save it to a file

stream.download(filename='audio')
```

```python
# We inspect the 'audio' file to check it's type and verify it's integrity'

!file audio
```

```python
# We inport the librosa library

import librosa.display

# IMporting the matplotlib library

import matplotlib.pyplot as plt

# We can use the librosa library to load the audio file

audio, sample_rate = librosa.load('audio')

# We can use the librosa library to plot the audio file

plt.figure(figsize=(14, 5))
librosa.display.waveshow(audio, sr=sample_rate)

```

```python
# To save the audio file we can use the librosa library

import librosa.output

librosa.output.write_wav('audio.wav', audio, sample_rate)
```
